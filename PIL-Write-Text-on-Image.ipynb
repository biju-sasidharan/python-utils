### Write Text on Image - Using PIL

This is an utility function to write text on images using Python imaging library. This utility is used to modify README.md file in all github repositories.

```python
# Load Libraries

import os

from PIL import Image, ImageFont, ImageDraw # Python imaging library
```

```python
# Constants

PATH = 'images/' # Images folder
INPUT_FILE_NAME = 'FIFA.jpg' # Inpput filename
OUTPUT_FILE_NAME = 'FIFA-PIL.jpg' # Inpput filename


TEXT = 'FIFA Worldcup' # text
FONT = ImageFont.truetype('fonts/PlayfairDisplay-Regular.ttf', 40) # font
POSITION = (0, 50) # position
COLOR = (0, 0, 255) # Text color
```

```python
img = Image.open(os.path.join(PATH, INPUT_FILE_NAME)) # Open image

img_editable = ImageDraw.Draw(img) # Provides 2D graphics on image object
img_editable.text(POSITION, TEXT, fill=COLOR, font=FONT) # Writing text to image

img.save(os.path.join(PATH, OUTPUT_FILE_NAME)) # Save  image 
img
```
